```python
%%bash

nohup vllm serve Qwen/Qwen2.5-1.5B-Instruct \
  --host 0.0.0.0 \
  --port 8000 \
  --served-model-name qwen2.5-1.5b \
  --dtype half \
  --max-model-len 2048 \
  --gpu-memory-utilization 0.85 \
  > /content/vllm_server.log 2>&1 &

echo $! > /content/vllm_server.pid
echo "Server process started with PID $(cat /content/vllm_server.pid)"
```

```python
import time
import requests

url = "http://127.0.0.1:8000/v1/models"

for attempt in range(90):
    try:
        response = requests.get(url, timeout=2)

        if response.status_code == 200:
            print("Server is ready.")
            print(response.json())
            break

    except requests.RequestException:
        pass

    if attempt % 10 == 0:
        print(f"Waiting for server... attempt {attempt + 1}")

    time.sleep(2)

else:
    print("Server did not become ready.")
    print("\nLast part of server log:\n")
    with open("/content/vllm_server.log", "r") as log_file:
        print(log_file.read()[-5000:])
```
